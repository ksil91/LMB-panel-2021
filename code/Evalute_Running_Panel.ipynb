## ID and remove bad samples

```sos
def p2f(x):
    return float(x.strip('%'))/100

db52_sample = inxl52.parse('Sample_Summary', converters={'CallRate':p2f})
#db52_sample.loc[100:,]
```

```sos
failed52_sampleDF = db52_sample[db52_sample['CallRate'] < 0.8]
failed52_sample = set(failed52_sampleDF["SAMPLE_NAME"] + "_"+failed52_sampleDF["WELL"])
failed52_sample
```

```sos
def stripX(x):
    return x[1:]

db39_sample = inxl39.parse('Sample_Summary', converters={'CallRate':p2f,'SAMPLE_NAME':stripX})
#db39_sample.loc[50:100,]
```

```sos
failed39_sampleDF = db39_sample[db39_sample['CallRate'] < 0.8]
failed39_sample = set(failed39_sampleDF["SAMPLE_NAME"] + "_"+failed39_sampleDF["WELL"])
failed39_sample
```

```sos
#Intersection
failed39_sample.intersection(failed52_sample)
```

```sos
# Union
union_bad_samp = set(failed52_sampleDF['SAMPLE_NAME']).union(set(failed39_sampleDF['SAMPLE_NAME']))
union_bad_samp
```

```sos
pass52 = db52_sample[~db52_sample['SAMPLE_NAME'].isin(union_bad_samp)]
pass39 = db39_sample[~db39_sample['SAMPLE_NAME'].isin(union_bad_samp)]
```

```sos
goodSamps = goodSamps + list(pass52['SAMPLE_NAME'])
```

```sos
# add to good Samps file
# add to existing goodSamps file
with open("goodSamps_0417_0529", "a") as myfile:
    myfile.writelines("%s\n" % s for s in sorted(goodSamps))
```
