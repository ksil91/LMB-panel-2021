```python
# convert vcf to genind in R, the read in loci names
lines = []
with open('./maf05mac4het9m3R50_2/Analysis/popgen/Hybrid_sim/m85-hwPBi-fixed3_98-thinned_loci.tab') as f:
    for line in f:
        line = line.strip().split(".")[0]
        line = tuple(map(int,line.split("__")))
        lines.append(line)
```

```python
x = INpure[INpure[["Chr","# Locus ID","BP"]].apply(tuple, 1).isin(lines)]
x = x[~x['Pop ID'].isin(["DBo","NBo","ILMB"])]
```

```python
ldict = {}
for index, row in x.iterrows():
    locus = str(row['Chr'])+ '__'+str(row['# Locus ID'])+'__'+str(row['BP'])
    if locus not in ldict.keys():
        ldict[locus]={row['Pop ID'] : row['P']}
        if 'P' not in ldict[locus].keys():
            if row['P Nuc'] != "-":
                ldict[locus]['P'] =row['P Nuc']
        if 'Q' not in ldict[locus].keys():
            if row['Q Nuc'] != "-":
                ldict[locus]['Q'] =row['Q Nuc']
    else:
        if row['Pop ID'] not in ldict[locus].keys():
            ldict[locus][row['Pop ID']] = row['P']
            if 'P' not in ldict[locus].keys():
                if row['P Nuc'] != "-":
                    ldict[locus]['P'] =row['P Nuc']
            if 'Q' not in ldict[locus].keys():
                if row['Q Nuc'] != "-":
                    ldict[locus]['Q'] =row['Q Nuc']
```
